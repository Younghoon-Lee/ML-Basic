> Gradient Descent

- 1차미분한 값을 사용해서 계속 반복적으로 최적화 시키고 local minimum 으로 최적화됨 

> Important Concepts in Optimization

- Generalization
- Under-fitting vs over-fitting
- Cross validation
- Bias-variance tradeoff
- Bootstrapping
- Bagging and boosting


> Generalization



- traing error 가 0이 되었다고 해서 항상 최적값을 보장하다는 법은 없음
- 어느정도 iteration이 지나가면 오히려 test에서 성능이 떨어짐 
- 흔히, 좋은 generalization performance를 가진다고 하면 학습데이터와 테스트데이터가 비슷한 성능을 갖는다는 말



> Underfitting vs Overfitting 



- underfitting : 네트워크가 너무 간단하거나 트레이닝 횟수가 적어서 학습데이터 조차 최적화 하지 못함
- overfitting : 과적합



> Cross-validation



- 기본적인 컨셉은 학습데이터(training dataset) 안에서 검증데이터를 따로 분류해서 검증
- cross-validation은 train data를 여러개로 파티셔닝후 각각의 조합으로 여러 경우의 수의 검증을 거침

> Bias and Variance 



- variance : 내가 어떤 입력을 넣었을 때 출력이 얼마나 일관적으로 나오는지의 정도 (낮을수록 일관적임)
- bias : 출력 값들의 mean이 target 값과 얼마나 비슷한지 (낮을수록 비슷함)


> Bias and Varainace Tradeoff



- 학습데이터에 noise가 껴있을 경우 bias와 variance를 둘다 줄이는 것은 사실상 힘들다는 원리 

> Bootstrapping

- any test or metric that uses random sampling with replacement



> Baggin vs Boosting 

- Bagging(Bootstrapping aggregating)
    - 모델을 만들 때, 갖고 있는 모든 train dataset을 사용하는 것이 아닌 일부의 train data를 샘플링하여 여러 모델을 만든 후 평균을 내어 최적화를 하는 작업 

- Boosting
    - 모델들을 여러개 학습시켜 하나의 큰 strong learner을 만드는 작업 
    - 일반적으로 특정 데이터에 있어서 학습이나 분류가 잘 안될 때 따로따로 모델을 만들어서 이어붙이는 것을 지칭 


> **Gradient Descent Methods**

- stochasitc gradient descent
    - 한개의 샘플의 gradient를 이용해서 update
    
- Mini-batch gradient descent
    - 128, 256개의 샘플의 gradinet를 한번에 이용해서 update

- Batch gradient descent
    - 한번의 모든 샘플의 gradient 평균을 이용해서 update
