>**Linear Neural Networks**

- Data: $D ={(x_i, y_i)}_{i=1}^N$

- Model: $\hat{y} = wx + b$

- Loss: $loss = \frac{1}{N}\sum_{i=1}^{N}(y_i-\hat{y}_i)^2$
