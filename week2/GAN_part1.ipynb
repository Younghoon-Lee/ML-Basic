> Learning a Generative Model

- 강아지 사진이 있다고 가정해보자

- we want to learn a probability distribution $p(x)$
  - generation: If we sample $x_{new} ~ p(x), x_{new}$ should look like a dog (sampling)
  - Density estimation: $p(x)$ should be high if $x$ looks like a dog, and low otherwise(anomaly detection)
  - 엄밀한 의미에 Generative model 은 생성과 분류 둘다 가능
    - explicit generation model: 생성, 분류 모두 가능 / implicit generation model : 생성만 가능
  - Unsupervised representation learning: We should be able to learn what these images have in common (feature learning)
  

> Basic Discrete Distributions

- Bernoulli distribution: (biased) coin flip
  - $D = \{Heads, Tails\}$
  - specify $P(X = Heads) = p$. Then $P(X=Tails) = 1 -p$ 
    - 즉, 확률이 하나만 정의 되어도 됨 $p$ , 나머지 확률은 $1-p$로 표현
  - Write : $X \sim Ber(p)$.
- Categorical distribution: (biased) m-sided dice
  - $D= \{1,...,m\}$
  - specify $P(Y=i)=p_i$, such that $\sum_{i=1}^{m}pi =1$
    - 확률이 i-1 개만 정의되고 나머지 확률은 1에서 뺀 나머지로 정의
  - Write : $Y \sim Cat(p_1,...,p_m)$
